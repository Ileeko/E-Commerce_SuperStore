### Encoding function for reading dataset:

Create a function they try witch encoding work to read our data :

```python
# Creating a function that give us witch encoding can we use to read the csv file with pandas
def read_csv_with_encoding(filename):
    """
    Function to read a CSV file with different encoding options.
    
    Parameters:
        filename (str): The path to the CSV file.
        
    Returns:
        DataFrame or None: Returns DataFrame if reading is successful, else returns None.
    """
    encoding_list=['utf-8', 'latin1', 'iso-8859-1', 'windows-1252', 'latin2', 'utf-16le', 'utf-16be', 'ascii', 'cp1252']
    
    for encoding in encoding_list:
        try:
            df = pd.read_csv(filename, encoding=encoding)
            print(f"CSV file successfully read with encoding: {encoding}")
        except Exception as e:
            print(f"Failed to read CSV with encoding '{encoding}': {e}")
    
    return None
```

```python
# Creat a variable with the path or the csv file we want to import 
csv_file = "Original_dataset/Sample-Superstore.csv"

# Use the function to get witch encoding we can use
read_csv_with_encoding(csv_file)
```

## Fixe the incoherence of Product:

### Unique product IDs having multiple name:

```python
# Group by 'product_id' and count the number of unique 'name' values in each group
product_id_counts = fixed_df.groupby('product_id')['product_name'].nunique()

# Check if any product has more than one unique 'name' value (product ID)
products_with_multiple_ids = product_id_counts[product_id_counts > 1]

# If products_with_multiple_ids is empty, it means each product has only one ID
if products_with_multiple_ids.empty:
    print("Each product has only one product ID.")
else:
    print("The following product have multiple product IDs:")
    print(products_with_multiple_ids)
```

### Creating a dictionary of Product ID choosing the first Product Name:

```python
# Get the product IDs with multiple names
products_with_multiple_ids = product_id_counts[product_id_counts > 1].index

# Create dictionary mapping each product ID to its first product name
first_product_names = fixed_df[fixed_df['product_id'].isin(products_with_multiple_ids)].groupby('product_id')['product_name'].first().to_dict()

# Print the dictionary
display(first_product_names)

# Total numbers of Products ID and names
display(len(first_product_names))
```

## Drop the Order Number column and Order by order_date:

```python
# Order the Dataframe by 'order_date' in descending order
df = df.drop(columns=['order_number']).sort_values(by='order_date', ascending=False)
# Reset the index
df = df.reset_index(drop=True)
```

## Add the column 'order_number':

```python
# Add a new column 'order_number' with values counting the rows in descending order
df['order_number'] = range(len(df), 0, -1)

# Reorder the columns to have 'order_number' as the first column
df = df[['order_number'] + [col for col in df.columns if col != 'order_number']]
```
